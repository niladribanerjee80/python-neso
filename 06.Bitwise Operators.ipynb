# Introduction to Number System

**Number System**\
Represents a quantity through a set of numbers\
e.g. I have \\$100 in my wallet.\
Here \\$100 is in decimal.

We have the following types of number system

| Number System   | Base | digits   |
| -------------   | ---- | -------- |
| Decimal         | 10   | 0-9.     |
| Binary          | 2    | 0,1.     |
| Octal           | 8    | 0-7.     |
| Hexadecimal     | 16   | 0-9, A-F |

### Decimal -> Binary

```python
dec_val = 678
bin_val = bin(dec_val)
print(bin_val)
```

```python
# if we don't want the 0b
dec_val = 678
bin_val = format(dec_val,'b')
print(bin_val)
```

### Decimal -> Octal

```python
dec_val = 678
oct_val = oct(dec_val)
print(oct_val)
```

```python
# if we don't want the 0o
dec_val = 678
oct_val = format(dec_val,'o')
print(oct_val)
```

### Decimal -> Hexadecimal

```python
dec_val = 678
hex_val = hex(dec_val)
print(hex_val)
```

```python
# if we don't want the 0x
dec_val = 678
hex_val = format(dec_val,'x')
print(hex_val)
```

### Binary -> Decimal

```python
bin_val = '011011'
base = 2
dec_val = int(bin_val,base)
print(dec_val)
```

### Octal -> Decimal

```python
oct_val = "1237" # use only 0-7 as octal numbers
base = 8
dec_val = int(oct_val,base)
print(dec_val)
```

### Hexadecimal -> Decimal

```python
hex_val = "12ff"
base = 16
dec_val = int(hex_val,base)
print(dec_val)
```

# Introduction to Bitwise Operator

__Definition : Bitwise operators operates on bits => binary digits.__

1. Bitwise AND = &
2. Bitwise OR  = |
3. Bitwise NOT = (~) # tilda
4. Bitwise XOR = (^) # Circumflex symbol

__1. Bitwise AND = &__ \
__Returns 1 if both bits are 1 or else returns 0__ \
&emsp;1 & 1 = 1 \
&emsp;1 & 0 = 0 \
&emsp;0 & 1 = 0 \
&emsp;0 & 0 = 0

__2. Bitwise OR  = |__ \
__Returns 1 if any bits is 1 or else returns 0__ \
&emsp;1 | 1 = 1 \
&emsp;1 | 0 = 1 \
&emsp;0 | 1 = 1 \
&emsp;0 | 0 = 0

__3. Bitwise NOT = ~__ \
__Returns 1's compliment of a number but using sign-bit__ \
Explanation required

__4. Bitwise XOR = (^)__ \
__Returns 1 if one of the bits is 1 or else returns 0__ \
&emsp;1 ^ 1 = 0 \
&emsp;1 ^ 0 = 1 \
&emsp;0 ^ 1 = 1 \
&emsp;0 ^ 0 = 0
    

```python
# Bitwise AND

# returns 1 if both bits are 1 or else returns 0

a = 10
b = 5

# a  = 1010
# b  = 0101
# ----------
#      0000
# ----------

a & b # the result is decimal
```

```python
# what if I want to see the binary code
bin(a & b)
```

```python
# Bitwise OR (|)

# Returns 1 if any bits is 1 or else returns 0

a = 10
b = 5

# a  = 1010
# b  = 0101
# ----------
#      1111
# ----------

a | b # the result is decimal
```

```python
# if you want to see the binary equivalent
bin (a | b)
```

### How Bitwise NOT works

Bitwise NOT returns 1's compliment

if a = 10 => 1010 \
Therefore, ~ a ideally should be 0101 which is 5. But let's see what is the result we get

```python
a = 10
~a
```

__Why are we getting -11__? \

a = 10 => (0)1010 \
We have a signed bit as well \

Signed Bits \

*   0 = +1
*   1 = -1

Now, 1's compliment of (0)1010 = (1)0101 \
=> -(2^4)\*1 + (2^3)\*0 + (2^2)\*1 + (2^1)\*0 + (2^0)\*1
= -11


```python
# Bitwise XOR = (^)
# Returns 1 if one of the bits is 1 or else returns 0

a = 10
b = 9

# a  = 1010
# b  = 1001
# ----------
#      0011
# ----------

a ^ b # the result is decimal
```

# Bitwise Shift Operations



*   Bitwise right shift (>>)
*   Bitwise left shift (<<)


**Bitwise right shift (>>)** \
Shifts the binary bits towards right. As a consequence, it divides the number by (2**n), where n = magnitude of the shift \

**Bitwise left shift (>>)** \
Shifts the binary bits towards left. As a consequence, it multiples the number by (2**n), where n = magnitude of the shift \

```python
# Bitwise right shift

a = 12 # this means (0)1100 in binary

# a >> 1 results in :
# (0) 1 1 0 0 = 12
#  0  0 1 1 0 = 6 ( 12 / 2 = 6)

a >> 1
```

```python
# a >> 2 means (a >> 1) >> 1:
#  0  0 1 1 0 = 6
#  0  0 0 1 1 = 3 (12 / 2**2 = 3)

a >> 2
```
